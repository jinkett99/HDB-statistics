#### Demand as counts per district code

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_normalized_counts(df, group_var):
    """
    Function to group a DataFrame by a specified variable, normalize counts, 
    and plot the normalized counts sorted by the grouping variable.

    Parameters:
    df (pd.DataFrame): The DataFrame to be analyzed.
    group_var (str): The column name of the grouping variable.
    """
    # Convert the grouping variable to string
    df[group_var] = df[group_var].astype(str)

    # Group by the grouping variable and obtain counts
    counts = df[group_var].value_counts()

    # Normalize the counts by total count
    normalized_counts = counts / counts.sum()

    # Sort normalized counts by the index (grouping variable) in ascending order
    normalized_counts = normalized_counts.sort_index()

    # Plotting
    plt.figure(figsize=(10, 6))
    normalized_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Normalized Counts of {group_var}')
    plt.xlabel(group_var)
    plt.ylabel('Normalized Count')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
# Example usage:
plot_normalized_counts(df, 'district')
```
